A] Write a python programme which display positive and negative numbers of a list into two separate lists and raises an error at the end of the programme.

```python
NumList = []
Positive = []
Negative = []
j = 0

Number = int(input("Please enter the Total Number of List Elements : "))
for i in range(1, Number + 1):
    value = int(input("Please enter the Value of %d Element : " %i))
    NumList.append(value)

while(j < Number):
    if(NumList[j] >= 0):
        Positive.append(NumList[j])
    else:
        Negative.append(NumList[j])
    j = j + 1

print("Element in Positive List is : ", Positive)
print("Element in Negative List is : ", Negative)
raise Exception("list is separated")
```
